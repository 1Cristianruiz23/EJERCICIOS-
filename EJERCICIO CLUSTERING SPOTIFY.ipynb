```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Ajustes visuales
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Cargar el dataset
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/MyDrive/DATASETS/spotify tracks db/SpotifyFeatures.csv")
# Mostrar las primeras filas
df.head()
```

```python
# Ver tipos de datos
df.info()

# Seleccionar columnas numéricas útiles para clustering
numeric_cols = [
    'acousticness', 'danceability', 'duration_ms', 'energy',
    'instrumentalness', 'liveness', 'loudness', 'speechiness',
    'tempo', 'valence', 'popularity'
]
df_numeric = df[numeric_cols]
df_numeric.describe()
```

```python
# Normalización de datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)
```

```python
# Aplicar PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Visualizar la varianza explicada
print(f"Varianza explicada: {pca.explained_variance_ratio_}")

# Crear DataFrame para visualización
pca_df = pd.DataFrame(pca_result, columns=["PCA1", "PCA2"])
pca_df["genre"] = df["genre"]

# Graficar
sns.scatterplot(data=pca_df, x="PCA1", y="PCA2", hue="genre", legend=False, alpha=0.5)
plt.title("PCA - Distribución por Género")
plt.show()
```

```python
# 1. Seleccionar columnas numéricas
numeric_cols = [
    'acousticness', 'danceability', 'duration_ms', 'energy',
    'instrumentalness', 'liveness', 'loudness', 'speechiness',
    'tempo', 'valence', 'popularity'
]

# 2. Tomar una muestra aleatoria de 5000 canciones
sample_df = df.sample(n=5000, random_state=42)
sample_genre = sample_df["genre"]  # guardamos el género para luego

# 3. Escalar los datos
scaler = StandardScaler()
scaled_sample = scaler.fit_transform(sample_df[numeric_cols])

# 4. Aplicar t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=500)
tsne_result = tsne.fit_transform(scaled_sample)

# 5. Crear DataFrame para graficar
tsne_df = pd.DataFrame(tsne_result, columns=["TSNE1", "TSNE2"])
tsne_df["genre"] = sample_genre.values

# 6. Graficar resultados
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tsne_df, x="TSNE1", y="TSNE2", hue="genre", legend=False, alpha=0.5)
plt.title("t-SNE (Muestra de 5000 canciones) - Distribución por Género")
plt.xlabel("Componente t-SNE 1")
plt.ylabel("Componente t-SNE 2")
plt.show()
```

```python
# Elegir número de clusters
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(pca_result)

# Agregar a DataFrame
pca_df["Cluster"] = clusters

# Visualizar clusters
sns.scatterplot(data=pca_df, x="PCA1", y="PCA2", hue="Cluster", palette="tab10")
plt.title("Clustering con KMeans sobre PCA")
plt.show()
```
