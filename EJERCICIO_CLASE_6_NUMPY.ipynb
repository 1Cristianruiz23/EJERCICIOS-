```python
import numpy as np

# Crear un arreglo 5x3 con números aleatorios enteros entre 0 y 100
arreglo = np.random.randint(0, 101, size=(5, 3))

print("Arreglo 5x3 de números aleatorios:")
print(arreglo)

# Calcular la suma total de todos los elementos
suma_total = np.sum(arreglo)

print("\nSuma total de los valores del arreglo:")
print(suma_total)

# Calcular el promedio de cada columna (axis=0)
promedios_columnas = np.mean(arreglo, axis=0)

print("\nPromedio de cada columna:")
print(promedios_columnas)
```

```python
import numpy as np

# Crear una matriz 10x10 con enteros aleatorios del 0 al 9
matriz = np.random.randint(0, 10, size=(10, 10))

print("Matriz 10x10 de enteros aleatorios entre 0 y 9:")
print(matriz)

# Reemplazar todos los valores mayores a 5 por -1
matriz[matriz > 5] = -1

print("\nMatriz con valores > 5 reemplazados por -1:")
print(matriz)
```

```python
img_array = np.array(img)
print("Dimensiones (alto, ancho, canales):", img_array.shape)
```

```python
# Separar canales
rojo = img_array[:, :, 0]
verde = img_array[:, :, 1]
azul = img_array[:, :, 2]

# Canal rojo
print("\nCanal Rojo")
print("Min:", rojo.min())
print("Max:", rojo.max())
print("Promedio:", rojo.mean())

# Canal verde
print("\nCanal Verde")
print("Min:", verde.min())
print("Max:", verde.max())
print("Promedio:", verde.mean())

# Canal azul
print("\nCanal Azul")
print("Min:", azul.min())
print("Max:", azul.max())
print("Promedio:", azul.mean())
```
